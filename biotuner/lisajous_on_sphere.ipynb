```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, CustomJS
from bokeh.layouts import column
from bokeh.models.widgets import Slider
from math import pi

output_notebook()

def lissajous_curves_on_sphere(tuning, delta, n_points=300):
    num = tuning
    denom = np.arange(1, len(tuning) + 1)
    t = np.linspace(-pi, pi, n_points)

    x = np.sin(num * t[:, np.newaxis] + delta)
    y = np.sin(denom * t[:, np.newaxis])

    X = 2 * x / (1 + x**2 + y**2)
    Y = 2 * y / (1 + x**2 + y**2)
    Z = (-1 + x**2 + y**2) / (1 + x**2 + y**2)

    return X, Y, Z

tuning = np.array([1, 2, 3, 4, 5])
delta = pi / 2
X, Y, Z = lissajous_curves_on_sphere(tuning, delta)

source = ColumnDataSource(data={'x': X[:, 0], 'y': Y[:, 0], 'z': Z[:, 0]})

plot = figure()
plot.scatter('x', 'y', source=source, line_width=2, alpha=0.8)

slider = Slider(title="Lissajous Curve Index", value=0, start=0, end=len(tuning)-1, step=1)

callback = CustomJS(args=dict(source=source, X=X, Y=Y, Z=Z), code="""
    var index = cb_obj.value;
    source.data['x'] = X.map(row => row[index]);
    source.data['y'] = Y.map(row => row[index]);
    source.data['z'] = Z.map(row => row[index]);
    source.change.emit();
""")

slider.js_on_change('value', callback)

layout = column(plot, slider)
show(layout)
```
